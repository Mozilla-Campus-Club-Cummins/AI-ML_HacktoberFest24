```python
#Importing all the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading and reading the dataset
df = pd.read_csv("StudentsPerformance.csv")
```

```python
#printing the dataset
df
```

```python
#select all object type attribute columns
categorical_cols = df.select_dtypes(include='object').columns
categorical_cols
```

```python
#view all the possible date of each attribute
for i in categorical_cols:
    print(df[i].unique())
```

```python
#visualisation of test prep course attribute
count_test = df['test preparation course'].value_counts()
labels = df['test preparation course'].value_counts().index
plt.figure(figsize= (6,6))
plt.pie(count_test,labels=labels,autopct='%1.1f%%')
plt.legend(labels)
plt.show()
```

```python
#reducing no of attributes by taking an average of the scores
df['average_score']=(df['math score']+df['reading score']+df['writing score'])/3
df
```

```python
sns.scatterplot(x=df['average_score'],y=df['math score'],hue=df['gender'])
sns.scatterplot(x=df['average_score'],y=df['reading score'],hue=df['gender'])
df
```

```python
#non ordinal grouping
gender = {
    'male':1,
    'female':0
}
race = {
    'group A':0,
    'group B':1,
    'group C':2,
    'group D':3,
    'group E':4
}
df['gender']=df['gender'].map(gender)
df['race/ethnicity']=df['race/ethnicity'].map(race)
df
```

```python
level = {
    "bachelor's degree":0,
    'some college':1,
    "master's degree":2,
    "associate's degree":3,
    "high school":4,
    "some high school":5
}
df['parental level of education']=df['parental level of education'].map(level)
df
```

```python
#Convert categorical variable into dummy/indicator variables
df = pd.get_dummies(df,drop_first=True)
df
```

```python
x = df.drop(columns='average_score').values
x
x[0]
y = df['average_score'].values
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
#splitting training and testing dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
#initialising object and fitting data in model
model=RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
predictions
```

```python
#checking accuracy of model(99.72%)
print(r2_score(predictions,y_test))
```
